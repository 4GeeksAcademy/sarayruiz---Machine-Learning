```python
import matplotlib.pyplot as plt
import seaborn as sns

# Columnas categóricas clave para el análisis
cat_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact']

# Configuración de los gráficos
fig, axes = plt.subplots(len(cat_cols), 1, figsize=(10, 5 * len(cat_cols)))
plt.subplots_adjust(hspace=0.6) # Espacio entre gráficos

for i, col in enumerate(cat_cols):
    # Calcular la tasa promedio de suscripción ('y_encoded') por cada categoría
    avg_subs = df.groupby(col)['y_encoded'].mean().sort_values(ascending=False).reset_index()
    
    sns.barplot(
        data=avg_subs,
        x=col,
        y='y_encoded',
        ax=axes[i],
        palette='viridis'
    )
    axes[i].set_title(f'Tasa Promedio de Suscripción por {col.capitalize()}', fontsize=14)
    axes[i].set_ylabel('Tasa de Suscripción (Media)', fontsize=12)
    axes[i].set_xlabel(col.capitalize(), fontsize=12)
    #axes[i].tick_params(axis='x', rotation=45, ha='right') # Rotar etiquetas
    axes[i].grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
